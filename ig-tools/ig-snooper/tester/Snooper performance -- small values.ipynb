```python
import matplotlib.pyplot as plt
import numpy as np
from itertools import groupby
from math import sqrt, ceil
```

```python
with open('small_data.txt') as file:
    values = list(map(lambda line: tuple(float(s) for s in line.split()), file))
    values_dict = {k: list(v) for k, v in groupby(sorted(values, key=lambda t: t[2]), key=lambda t: t[2])}
```

```python
ml_wnd_min = min(values, key=lambda t: t[0])[0]
ml_wnd_max = max(values, key=lambda t: t[0])[0]
avg_wnd_min = min(values, key=lambda t: t[1])[1]
avg_wnd_max = max(values, key=lambda t: t[1])[1]

max_values = {err_id: max(map(lambda t: t[3 + err_id], values)) for err_id in range(5)}
```

```python
def get_plot_data(k, err_id):
    return np.array([[next(t_inner)[err_id + 3] for avg_wnd, t_inner in groupby(t_outer, key=lambda t: t[1])] for ml_wnd, t_outer in groupby(sorted(values_dict[k], key=lambda t: t[0]), key=lambda t:t[0])])
```

```python
def plot(err_id):
    ks = list(map(int, sorted(values_dict)))
    w = h = math.ceil(sqrt(len(ks)))
    w-= 1
    fig = plt.figure(figsize=(10, 10), tight_layout={'pad': 1.08, 'h_pad': 1, 'w_pad': 1, 'rect': (0, 0, .8, 1)})
    last_plot = None
    for i, k in enumerate(ks):
        plt.subplot(h, w, i + 1)
        plot_data = get_plot_data(k, err_id)
        plt.title('thresh={0}'.format(k))
        last_plot = plt.imshow(plot_data, interpolation='none', origin='lower', extent=[avg_wnd_min, avg_wnd_max, ml_wnd_min, ml_wnd_max], vmin=0, vmax=max_values[err_id])
        plt.xlabel('averaging window')
        plt.ylabel('learning window')
    cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
    plt.tick_params(axis='x', which='both', bottom='off', top='off', labelbottom='off')
    fig.colorbar(last_plot, cax=cbar_ax)
    plt.show()
```
